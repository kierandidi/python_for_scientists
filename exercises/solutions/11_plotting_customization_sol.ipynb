Try to reproduce the plot below as well as possible and save it as a .png file. The style-sheet used for the plot was the "ggplot" style-sheet. To get the colors right, you can use the HEX codes from Google's color picker (simply google "color picker") - it does not have to be perfect.

```python
import seaborn as sns
import matplotlib.pyplot as plt

df = sns.load_dataset("penguins")
df
```

```python
plt.style.use("ggplot")

plt.figure(figsize=(7.5, 4.8))

palette = ["#264653", "#2A9D8F", "#E9C46A"]
sns.scatterplot(
    data=df,
    x="flipper_length_mm",
    y="bill_length_mm",
    hue="species",
    style="species",
    markers=["H", "s", "D"],
    alpha=0.75,
    palette=palette,
    s=60,
)

plt.title("Flipper length vs. bill length for penguin species", fontsize=15)
plt.xlabel("Flipper length [mm]")
plt.ylabel("Bill length [mm]")

plt.legend(loc="center left", bbox_to_anchor=(1, 0.5), frameon=False)
plt.savefig("../../img/plots/scatterplot_ex.png", dpi=300, bbox_inches="tight")
```

```python
# this resets the style-sheet from your above code
plt.rcdefaults()
```

```python
# seaborn standard theme
sns.set_theme()

g = sns.relplot(
    data=df,
    x="flipper_length_mm",
    y="bill_length_mm",
    hue="species",
    style="species",
    col="island",
    col_wrap=2,
    aspect=6 / 5,
    height=3.2,
)

g.set_xlabels(label="Flipper length [mm]")
g.set_ylabels(label="Bill length [mm]")
plt.suptitle(
    "Flipper length vs. bill length for different penguin species by island",
    y=1.04,
    fontsize=16,
)

sns.move_legend(
    g,
    loc="center",
    bbox_to_anchor=(0.65, 0.28),
    title=None,
    fontsize=14,
    title_fontsize=14,
    markerscale=1.2,
)

plt.savefig("../../img/plots/gridplot_ex.png", dpi=300, bbox_inches="tight")
```
